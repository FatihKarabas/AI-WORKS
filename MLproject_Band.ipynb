## *info*

Hava Kalitesi İndeksi Nedir?
Tüm dünyada yaygın olarak kullanılan, Hava Kalitesi İndeksi (HKİ) olarak adlandırılan sınıflama sistemi ile kirleticilerin konsantrasyonlarına göre hava kalitesini iyi, orta, kötü, tehlikeli vb. şeklinde sınıflandırılması yapılmaktadır. Dünyanın pek çok ülkesinde indeks hesaplanmasında kullanılan yöntem ve kriterler, kendi ülkelerinde uygulanan hava kalitesi standartlarına uygun şekilde oluşturulmuştur.

Atmosferdeki hava kalitesini değerlendirmek için kullanılan HKİ, 0’dan 500’e kadar bir ölçekte yer alır. Bu ölçek, HKİ değeri ne kadar yüksekse, hava kirliliği seviyesinin o kadar büyük olduğunu ve buna bağlı olarak sağlık endişesinin arttığını gösterir. Örneğin, 50 HKİ değeri, halk sağlığını minimum düzeyde etkileyen iyi hava kalitesini temsil ederken, 300’ün üzerindeki bir HKİ değeri tehlikeli hava kalitesini ifade eder.

Hava Kalitesi İndeksi, renkler ve kategorilerle ifade edilir. Örneğin, yeşil renk ve 0-50 aralığı, iyi hava kalitesini; sarı renk ve 51-100 aralığı, orta düzeyde hava kalitesini; kırmızı renk ve 101-150 aralığı, hassas gruplar için sağlıksız hava kalitesini temsil eder. Bu renkli kategoriler, halkın hava kalitesini kolayca anlamasına ve gerektiğinde önlemler almasına yardımcı olur.

 İyi (0-50)
Hava kalitesi memnun edici ve hava kirliliği az riskli veya hiç risk teşkil etmiyor.

 Orta (51-100)
Hava kalitesi uygun fakat alışılmadık şekilde hava kirliliğine hassas olan çok az sayıdaki insanlar için bazı kirleticiler açısından orta düzeyde sağlık endişesi oluşabilir.

 Hassas (101-150)
Hassas gruplar için sağlık etkileri oluşabilir. Genel olarak kamunun etkilenmesi olası değildir.

 Sağlıksız (151-200)
Herkes sağlık etkileri yaşamaya başlayabilir, hassas gruplar için ciddi sağlık etkileri söz konusu olabilir.

 Kötü (201-250)
Sağlık açısından acil durum oluşturabilir. Nüfusun tamamının etkilenme olasılığı yüksektir.

 Tehlikeli (251-300)
Sağlık alarmı: Herkes daha ciddi sağlık etkileri ile karşılaşabilir.

### *info*

🔬 Kirletici Değişkenler ve Açıklamaları
1. PM10 (Partikül Madde 10 mikron)
Tanım: 10 mikrometreden küçük çaplı katı veya sıvı parçacıkla (toz, kül, polen vb.)

Kaynaklar: Endüstriyel faaliyetler, taşıt egzozları, inşaat çalışmaları, yakıt yakma.

Sağlık Etkileri: Akciğerlere kadar inebilir; astım, bronşit ve kalp-damar hastalıklarını tetikler. Uzun süreli maruziyet ölümlere neden olabilir.

Not: Dünya Sağlık Örgütü (DSÖ) PM10 için yıllık ortalama sınır: 15 µg/m³ (2021 güncellemesi)

2. SO₂ (Kükürt Dioksit)
Tanım: Fosil yakıtların (özellikle kömür ve petrol) yanmasıyla oluşan renksiz, keskin kokulu bir gaz.

Kaynaklar: Termik santraller, sanayi bacaları, gemi taşımacılığı.

Sağlık Etkileri: Solunum yollarını tahriş eder; kısa sürede maruziyet öksürük, göğüs sıkışması yapabilir. Astım hastalarında özellikle tehlikelidir.

Çevresel Etki: Asit yağmurlarının ana nedenidir.

3. NO₂ (Azot Dioksit)
Tanım: Azot oksitlerinin (NOx) bir formu; kahverengi renkli, keskin kokulu bir gazdır.

Kaynaklar: Araç egzozları, termik santraller, fabrikalar.

Sağlık Etkileri: Kronik akciğer hastalıkları ve enfeksiyonlara yatkınlık oluşturur. Çocuklarda akciğer gelişimini olumsuz etkiler.

İkincil Kirletici Rolü: Ozon (O₃) ve ince partiküllerin (PM2.5) oluşumuna katkıda bulunur.

4. NOx (Azot Oksitler – NO + NO₂)
Tanım: NO (Azot monoksit) ve NO₂ birleşiminden oluşan toplam gaz grubu.

Kaynaklar: Motorlu taşıtlar, sanayi tesisleri, enerji santralleri.

Sağlık Etkileri: NO tek başına nispeten zararsızdır, ancak havadaki oksijenle birleşip NO₂'ye dönüşür.

İkincil Etki: Fotokimyasal duman (smog) ve troposferik ozonun oluşumuna katkı sağlar.

5. NO (Azot Monoksit)
Tanım: Yanma süreçlerinde ortaya çıkan renksiz, kokusuz bir gaz.

Kaynaklar: Özellikle taşıtlar ve endüstriyel prosesler.

Sağlık Etkileri: NO kendi başına çok zararlı değildir, ancak havada kolaylıkla NO₂’ye oksitlenerek dolaylı zarara yol açar.

6. O₃ (Ozon – Troposferik)
Tanım: Yeryüzüne yakın seviyelerde oluşan ozon, kirletici gazlardan (özellikle NOx ve VOC) güneş ışığıyla etkileşerek oluşur.

Kaynaklar: Kendisi doğrudan salınmaz; ikincil kirleticidir.

Sağlık Etkileri: Solunum sistemine zarar verir, özellikle çocuklar ve yaşlılar risk altındadır.

Not: Atmosferin üst tabakasındaki ozon zararlı UV ışınlarına karşı koruyucudur; ancak yeryüzüne yakın ozon insan sağlığı için zararlıdır.

Aykırı Değer Analizi

````python
def aykiri_istatistik(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    alt = Q1 - 1.5 * IQR
    ust = Q3 + 1.5 * IQR
    aykiri_mask = (df[col] < alt) | (df[col] > ust)
    aykiri_sayi = aykiri_mask.sum()
    toplam = df[col].count()
    oran = 100 * aykiri_sayi / toplam
    return aykiri_sayi, oran

kolonlar = df.select_dtypes(include=['float64', 'int64']).columns
sonuc = []

for col in kolonlar:
    sayi, oran = aykiri_istatistik(df, col)
    if sayi > 0:
        sonuc.append({'Değişken': col, 'Aykırı Sayı': sayi, 'Oran (%)': round(oran, 2)})

# DataFrame olarak yazdır
aykiri_df = pd.DataFrame(sonuc).sort_values(by='Oran (%)', ascending=False)
print(aykiri_df.to_string(index=False))
````

````python
def aykiri_flagla(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    alt = Q1 - 1.5 * IQR
    ust = Q3 + 1.5 * IQR
    df[f'{col}_aykiri'] = ((df[col] < alt) | (df[col] > ust)).astype(int)
    return df

for col in out_cols:
    df = aykiri_flagla(df, col)
````

# **📊 Hava Kalitesi İndeksi (HKİ)**

HKİ Hesaplama Formülü:

HKİ, belirli bir kirleticinin ölçülen konsantrasyonunun, o kirletici için tanımlanmış eşik değerler aralığında hangi HKİ kategorisine denk geldiğini belirlemek için aşağıdaki formül kullanılarak hesaplanır:

image.png

🧮 Uygulama Örneği: PM₁₀ için HKİ Hesaplaması
Örneğin, PM₁₀ için eşik değer aralıkları ve karşılık gelen HKİ değerleri aşağıdaki gibidir:


| PM₁₀ Konsantrasyonu (µg/m³) | HKİ Değeri |
| --------------------------- | ---------- |
| 0 – 50                      | 0 – 50     |
| 51 – 100                    | 51 – 100   |
| 101 – 150                   | 101 – 150  |
| 151 – 200                   | 151 – 200  |
| 201 – 300                   | 201 – 300  |
| 301 – 500                   | 301 – 500  |


# **Özellik Seçimi ve Dönüşümle**

````python
# Adım 1: Gerekli tanımlamalar
pol_cols_wins = ['pm10', 'so2_win', 'no2', 'nox', 'no', 'o3']

# AQI skalalarını Türkiye ulusal hava kalitesi standartlarına göre tanımlayalım
turkey_aqi_breakpoints = {
    'pm10': [(0, 50, 0, 50), (51, 100, 51, 100), (101, 260, 101, 150),
             (261, 400, 151, 200), (401, 520, 201, 300), (521, 1000, 301, 500)],
    'so2_win': [(0, 100, 0, 50), (101, 250, 51, 100), (251, 500, 101, 150),
                (501, 850, 151, 200), (851, 1100, 201, 300), (1101, 2000, 301, 500)],
    'no2': [(0, 100, 0, 50), (101, 200, 51, 100), (201, 500, 101, 150),
            (501, 1000, 151, 200), (1001, 2000, 201, 300), (2001, 3000, 301, 500)],
    'nox': [(0, 100, 0, 50), (101, 200, 51, 100), (201, 500, 101, 150),
            (501, 1000, 151, 200), (1001, 2000, 201, 300), (2001, 3000, 301, 500)],
    'no': [(0, 100, 0, 50), (101, 200, 51, 100), (201, 500, 101, 150),
           (501, 1000, 151, 200), (1001, 2000, 201, 300), (2001, 3000, 301, 500)],
    'o3': [(0, 120, 0, 50), (121, 160, 51, 100), (161, 180, 101, 150),
           (181, 240, 151, 200), (241, 700, 201, 300), (701, 1000, 301, 500)]
}

# AQI hesaplama fonksiyonu
def calculate_aqi(value, breakpoints):
    for c_low, c_high, i_low, i_high in breakpoints:
        if c_low <= value <= c_high:
            return ((i_high - i_low) / (c_high - c_low)) * (value - c_low) + i_low
    return 500

# Adım 2: Ortalama hesaplamaları (pm10 için 24 saatlik, o3 için 8 saatlik)
df['pm10_24h'] = df['pm10'].rolling(window=24, min_periods=1).mean()
df['o3_8h'] = df['o3'].rolling(window=8, min_periods=1).mean()

# Adım 3: AQI sütunlarını oluştur
aqi_cols = {}
for col in pol_cols_wins:
    if col == 'pm10':
        avg_col = 'pm10_24h'
    elif col == 'o3':
        avg_col = 'o3_8h'
    else:
        avg_col = col
    aqi_col = col + '_AQI'
    df[aqi_col] = df[avg_col].apply(lambda x: calculate_aqi(x, turkey_aqi_breakpoints[col]))
    aqi_cols[col] = aqi_col

# Adım 4: Gen_AQI sütunu: satır bazında maksimum AQI
df['Gen_AQI'] = df[list(aqi_cols.values())].max(axis=1)

#import ace_tools as tools; tools.display_dataframe_to_user(name="Genel AQI ve Kirletici Bazlı AQI", dataframe=df)
print(df[['pm10', 'pm10_24h', 'pm10_AQI', 'Gen_AQI']].head())
````

````python
# Gen_AQI değerlerini Türkiye Hava Kalitesi Kategorilerine göre sınıflandıralım

def classify_aqi(aqi):
    if aqi <= 50:
        return "İyi"
    elif aqi <= 100:
        return "Orta"
    elif aqi <= 150:
        return "Hassas"
    elif aqi <= 200:
        return "Zararlı"
    elif aqi <= 300:
        return "Çok Zararlı"
    else:
        return "Tehlikeli"

# Yeni kategorik sütunu oluştur
df['Gen_AQI_Kategori'] = df['Gen_AQI'].apply(classify_aqi)

import pandas as pd
# Sonuçları göster
df_out = df[['Gen_AQI', 'Gen_AQI_Kategori'] + [col + '_AQI' for col in pol_cols_wins]]
print(df_out.head())
````
